5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude
establishments.update_many({}, [ {'$set':{ "geocode.longitude" : {'$toDecimal': "$geocode.longitude"}}} ])
```

```python
# Change the data type from String to Decimal for latitude
establishments.update_many({}, [ {'$set':{ "geocode.latitude" : {'$toDecimal': "$geocode.latitude"}}} ])
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many({}, [ {'$set':{ "RatingValue" : {'$toInt': "$RatingValue"}}} ])
```

```python
# Check that the geocode.longitude type is now a number
pipeline_long = [
    {"$project": {"field_type": {"$type": "$geocode.longitude"}}},
    {"$limit": 1}
]
result_long = establishments.aggregate(pipeline_long)

# Print the result
for doc in result_long:
    print("The data type of the geocode.longitude field is:", {doc['field_type']})
```

```python
# Check that the geocode.latitude type is now a number
pipeline_lat = [
    {"$project": {"field_type": {"$type": "$geocode.latitude"}}},
    {"$limit": 1}
]
result_lat = establishments.aggregate(pipeline_lat)

# Print the result
for doc in result_lat:
    print("The data type of the geocode.latitude field is:", {doc['field_type']})
```

```python
# Check that the RatingValue type is now a number
pipeline_rv = [
    {"$project": {"field_type": {"$type": "$RatingValue"}}},
    {"$limit": 1}
]
result_rv = establishments.aggregate(pipeline_rv)

# Print the result
for doc in result_rv:
    print("The data type of the RatingValue field is:", {doc['field_type']})
```
